```python
import requests
from bs4 import BeautifulSoup

# URL of the LinkedIn job posting
url = "https://www.linkedin.com/jobs/view/3943030640" #Nordic 
url = "https://www.linkedin.com/jobs/view/3910639312" #Microsoft - Senior Applied Scientist
# Fetch the webpage content
response = requests.get(url)
web_content = response.text

# Initialize BeautifulSoup
soup = BeautifulSoup(web_content, 'html.parser')

# Function to extract the desired sections from the HTML
def extract_job_details(soup):
    job_details = {}

    # Extracting the text within the specified div element
    description_div = soup.find('div', {'class': 'description__text description__text--rich'})
    if description_div:
        job_details['Description Text'] = description_div.get_text(strip=True)

    return job_details

# Extract the job details using the function
job_details = extract_job_details(soup)

# Display the extracted job details
for section, text in job_details.items():
    print(f"{section}:\n{text}\n")
```
